```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_profiling
import numpy as np
```

```python
mj = pd.read_csv('https://raw.githubusercontent.com/tt-dspt-200/best-buds/main/data/MJ.csv')
```

```python
mj.head()
```

```python
mj['prop'] = mj['Ailment'] + ',' + mj['Effects_x'] + ',' + mj['Effects_y']
```

```python
mj.sample(7)
```

```python
type(mj['prop'][0])
```

```python
mj['prop'][7]
```

```python
import spacy 
from nltk.stem.snowball import SnowballStemmer
import re
```

```python
sp = spacy.load('en_core_web_lg')
```

```python
# 1. Remove non-alphanumeric characters
mj['prop'] = mj['prop'].apply(lambda x: re.sub('[^a-zA-Z]', ' ', x))

# 2. Remove extra whitespace 
mj['prop'] = mj['prop'].apply(lambda x: ' '.join(x.split()))

# Create Tokens
mj['tokens']= mj['prop'].apply(lambda x: sp.tokenizer(x))

# Find lemmas
mj['lemmas'] = ""
for i in range(len(mj)):
  for word in mj['tokens'][i]:
    if mj['lemmas'][i] == "":
      mj['lemmas'][i] = mj['lemmas'][i] + word.lemma_
    else:
      mj['lemmas'][i] = mj['lemmas'][i] + " " + word.lemma_
```

```python
mj['lemmas'][7]
```

```python
# Create Lemmatized Tokens
mj['tokens']= mj['lemmas'].apply(lambda x: sp.tokenizer(x))
```

```python
mj['tokens'][7]
```

```python
# Create vectors for model training
```

```python
for word in sp.tokenizer('depressed depression anxiety anxious'):
  print(word.text,  word.lemma_)
```

```python
sb = SnowballStemmer("english")
```

```python
for word in mj['tokens'][7]:
  print(word.text,  sb.stem(word.text))
```
